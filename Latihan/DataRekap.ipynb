```python
# mapping data {tahun: penjualan}
data_map = {data["tahun"]: data["penjualan"] for data in data_rekap}

data_map
```

```python
# sorting years
years = list(data_map.keys())
years.sort()
years
```

```python
# finalize data
data_graph = {}
for year in years:
    year_list = data_graph.get("years", [])
    sale_list = data_graph.get("sales", [])
    year_list.append(year)
    sale_list.append(data_map[year])
    data_graph["years"] = year_list
    data_graph["sales"] = sale_list
data_graph
```

```python
plt.plot(data_graph["years"], data_graph["sales"],'.--', label="Penjualan")
plt.plot(data_graph["years"], [i * 100 for i in range(len(years))],'--', label="Dummy")

plt.xlabel("Tahun")
plt.ylabel("Dalam jutaan rupiah")
plt.title("Data Rekapitulasi Tahunan")
plt.legend()
plt.show()
```

```python
# List untuk menyimpan hasil
income_data = []
profit_data = []

# Menghitung income dan profit per tahun
for data in data_rekap:
    tahun_income = data["penjualan"] + data["pendapatan_lain"]
    tahun_profit = tahun_income - data["operasional"] - data["pajak"]
    
    income_data.append({"tahun": data["tahun"], "income": tahun_income})
    profit_data.append({"tahun": data["tahun"], "profit": tahun_profit})

# Output hasil
print("Income per tahun:", income_data)
print("Profit per tahun:", profit_data)
```

```python
# Data awal dari data_rekap
data_graph = {
    "years": [],
    "sales": [],
    "income": [],
    "profit": []
}

# Mengisi data_graph dari data_rekap
for data in data_rekap:
    tahun_income = data["penjualan"] + data["pendapatan_lain"]
    tahun_profit = tahun_income - data["operasional"] - data["pajak"]
    
    data_graph["years"].append(data["tahun"])
    data_graph["sales"].append(data["penjualan"])
    data_graph["income"].append(tahun_income)
    data_graph["profit"].append(tahun_profit)

# Mengurutkan data berdasarkan tahun
sorted_data_graph = {
    "years": [],
    "sales": [],
    "income": [],
    "profit": []
}

# Menggabungkan semua data ke dalam list of tuples untuk sorting
combined_data = list(zip(data_graph["years"], data_graph["sales"], data_graph["income"], data_graph["profit"]))

# Sorting berdasarkan tahun (index 0)
combined_data.sort()

# Memisahkan kembali hasil sorting ke data_graph
for year, sale, inc, prof in combined_data:
    sorted_data_graph["years"].append(year)
    sorted_data_graph["sales"].append(sale)
    sorted_data_graph["income"].append(inc)
    sorted_data_graph["profit"].append(prof)

sorted_data_graph
```
